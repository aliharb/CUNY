First, lets clean up the data:

```python
import pandas as pd
riverkeeper = pd.read_csv('riverkeeper_data_2013.csv')
```

```python

riverkeeper.EnteroCount = riverkeeper.EnteroCount.str.replace('>','')
riverkeeper.EnteroCount = riverkeeper.EnteroCount.str.replace('<','')
```

```python
riverkeeper.EnteroCount = riverkeeper.EnteroCount.apply(int)
riverkeeper.FourDayRainTotal = riverkeeper.FourDayRainTotal.apply(int)
riverkeeper.SampleCount = riverkeeper.SampleCount.apply(int)
```

# Question 1

Create lists & graphs of the best and worst places to swim in the dataset.

```python
import seaborn as sns
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
groupedBySite = riverkeeper.groupby(riverkeeper['Site'], as_index=False).mean()

groupedBySite = groupedBySite.sort_index(by='EnteroCount', ascending=False).reset_index()
```

Top 10 dirtiest places to swim:

```python
groupedBySite[0:10]
```

Top 10 cleanest places to swim:

```python
groupedBySite.sort_index(by='EnteroCount')[0:10]

```

Graph of the top 10 places to swim, along with their EnteroCounts:

```python
#sns.factorplot("kind", "pulse", "diet", exercise, kind="point");
useddata = groupedBySite.sort_index(by='EnteroCount')[0:10]

sns.factorplot('Site', 'EnteroCount', data=useddata, aspect = 2.5, kind='bar', x_order=useddata['Site'].values).set_xticklabels(rotation=-30)
#useddata
```

```python
useddata = groupedBySite.sort_index(by='EnteroCount', ascending=False)[0:10]

sns.factorplot('Site', 'EnteroCount', data=useddata, aspect = 2.5, kind='bar', x_order=useddata['Site'].values).set_xticklabels(rotation=-30)
```

# Question 2

The testing of water quality can be sporadic. Which sites have been tested most regularly? Which ones have long gaps between tests? Pick out 5-10 sites and visually compare how regularly their water quality is tested.

```python
import datetime


sitelist = ['East River mid-channel at Roosevelt Is.', 'Esopus Creek West', 'Norrie Point Yacht Basin',
            'North River STP at 145th', 'Nyack Launch Ramp', 'The Battery mid-channel', 'Yonkers mid-channel']

Question2Data = riverkeeper[riverkeeper['Site'].isin(sitelist)]


Question2Data['Date'] = pd.to_datetime(Question2Data.loc[:,'Date']).values
```

```python
sns.set_palette("deep", desat=.6)

#useddata = Question2Data[Question2Data.Site == '125th St. Pier']

useddata = Question2Data

useddata['MonthYear'] = useddata.loc[:,'Date'].apply(lambda x : datetime.date(x.year, x.month, 1))



useddata = useddata.groupby(by=['Site','MonthYear'], as_index=False).sum()

#gammas = sns.load_dataset("gammas")
#gammas['time'] = pd.to_datetime('2000-01-01') + pd.to_timedelta(10, unit='D') * gammas['timepoint']
#sns.tsplot(gammas, "time", "subject", "ROI", "BOLD signal")

#sns.tsplot(useddata, 'MonthYear','SampleCount')

sns.factorplot(x='MonthYear', y='SampleCount', data=useddata, row='Site', aspect = 2.5, kind='bar').set_xticklabels(rotation=90)


```
