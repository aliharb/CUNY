```python
import plotly.plotly as py
import plotly.graph_objs as go
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
img = plt.imread('./Emotions/facial_expressions/images/images/angry1.png')
plt.imshow(img[:,:,2])
```

```python
data = [
    go.Heatmap(
        z = np.flipud(img[:,:,2])
    )
]

layout = go.Layout(
    height = 500,
    width = 500
)

py.iplot(go.Figure(data = data, layout = layout), filename = 'angry-man')
```
