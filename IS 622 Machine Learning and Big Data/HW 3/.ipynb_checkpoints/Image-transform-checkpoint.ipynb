```python
import cv2
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt

img = cv2.imread('./Emotions/facial_expressions/images/images/angry1.png')
```

These two functions create a random affine transformation of an image (greyscale or color). It takes in two parameters: the image itself, and a pctCov parameter. pctCov can be between 0 and 1, and is used to control the percentage of black space showing. Without this, I would get a lot of long and skinny transformations. I've been keeping this number pretty high, even at around 0.95, you get some images that aren't recognizable as a face (by this human)

```python
def randomize(img, pctCov):
    rows, cols = img.shape
    c11 = np.random.uniform(low = 0, high = 3)
    c21 = np.random.uniform(low = 0, high = 3)
    c12 = np.random.uniform(low = 0, high = 3-c11)
    c22 = np.random.uniform(low = 0, high = 3-c21)
    c13 = np.random.uniform(low = 0, high = 3-c11-c12)*350
    c23 = np.random.uniform(low = 0, high = 3-c21-c22)*350
    c23 = np.random.uniform(low = 0, high = 3-c21-c22)*350
    
    M = np.float32([[c11,c12,c13],[c21,c22,c23]])
    dst = cv2.warpAffine(img,M,(int(cols*3),int(rows*3)))
    Mreduce = np.float32([[0.3,0,0],[0,0.3,0]])
    dst = cv2.warpAffine(dst,Mreduce, (cols, rows))
    if float(dst[dst != 0].size)/float(dst.size) > (pctCov/9):
        return dst
    else:
        return randomize(img, pctCov)
    
def randomizeCol(img, pctCov):
    rows, cols = img[:,:,0].shape
    c11 = np.random.uniform(low = 0, high = 3)
    c21 = np.random.uniform(low = 0, high = 3)
    c12 = np.random.uniform(low = 0, high = 3-c11)
    c22 = np.random.uniform(low = 0, high = 3-c21)
    c13 = np.random.uniform(low = 0, high = 3-c11-c12)*350
    c23 = np.random.uniform(low = 0, high = 3-c21-c22)*350
    c23 = np.random.uniform(low = 0, high = 3-c21-c22)*350
    
    M = np.float32([[c11,c12,c13],[c21,c22,c23]])
    Mreduce = np.float32([[0.3,0,0],[0,0.3,0]])
    dst2 = img[:,:,0].reshape(rows, cols, 1)
    for i in np.arange(0,3):
        dst = cv2.warpAffine(img[:,:,i],M,(int(cols*3),int(rows*3)))
        dst = cv2.warpAffine(dst,Mreduce, (cols, rows))
        if i == 0:
            dst2 = dst.reshape(rows, cols, 1)
        else:
            dst2 = np.concatenate([dst2, dst.reshape(rows,cols,1)], axis = 2)
    if float(dst[dst != 0].size)/float(dst.size) > (pctCov/9):
        return dst2
    else:
        return randomizeCol(img, pctCov)
```

```python
plt.imshow(randomizeCol(img, 0.95))
```
